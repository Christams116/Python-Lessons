# üîÑ Conversion Functions: str(), int(), and float()
In Python, you sometimes need to change the "type" of a piece of information so you can use it correctly.

## ‚úÖ The str() Function: Changing to Words
The str() function turns any value into a string. This is very useful when you want to join a number to a sentence.

Example:

```python
age = 10
# We use str() to turn the number 10 into the string "10"
print("I am " + str(age) + " years old.")
```

## ‚úÖ The int() Function: Changing to Whole Numbers
The int() function takes a string or a decimal and turns it into a whole number (integer). This is how we fix the "input trap" where input() always gives us a string.

Example:

```python
# Convert the string "5" into the number 5
number_of_cats = int("5")
print(number_of_cats + 2)
```

## ‚úÖ The float() Function: Changing to Decimals
The float() function turns a value into a decimal number. Use this if you are working with money, weight, or anything that needs a . point.

Example:

```python
# Convert the string "1.5" into a decimal number
price = float("1.5")
print(price * 2)
```

## ‚ö†Ô∏è IMPORTANT: The "One-Way Street" Rule
You can only convert a string to a number if the string looks like a number.

int("10") works because "10" looks like a number ‚úÖ

int("Hello") will cause an error because Python doesn't know how to turn letters into math numbers ‚ùå

## üöÄ Pro Tip: Combining with input()
You can "wrap" the input() function inside a conversion function to get a number from a user immediately.

Example:

```python
# This asks for a number AND converts it to an int all at once!
score = int(input("What is your score? "))
print(score + 10)
```
