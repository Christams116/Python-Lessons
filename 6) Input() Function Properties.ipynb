# ‚å®Ô∏è Getting Input from the User: input() (Input Properties)

Sometimes we don‚Äôt want Python to just print things‚Ä¶ we want Python to ask the user questions and save their answers.

That‚Äôs what input() does!

## ‚úÖ Property 1: input() pauses the program

When Python sees input(), it will stop and wait until the user types something and presses Enter.

Example:

```python
color = input("What is your favorite color: ")
```

üí° **Note:** The words inside the quotes are called a prompt (it tells the user what to type).

## ‚úÖ Property 2: input() stores what the user typed

Whatever the user types gets saved into the variable.

Example:

```python
color = input("What is your favorite color: ")
print("Your favorite color is " + color)
```

## ‚úÖ Property 3: input() always gives you a string

#### **This is SUPER important:**

Even if the user types a number like 10‚Ä¶

```python
age = input("How old are you? ")

print(type(age))
```

Python saves it as a **string:**

"10" ‚úÖ (string)

not 10 ‚ùå (number)

üí° **Note:** If you want to do math with it later, you will need int() or float() (we‚Äôll learn that next).

## ‚úÖ Property 4: You can use multiple inputs

You can ask more than one question and store multiple answers.

Example:

```python
name = input("What is your name? ")
hobby = input("What is your favorite hobby? ")

print("Your name is " + name)
print("Your favorite hobby is " + hobby)
```
