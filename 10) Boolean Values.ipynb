# üö¶ Boolean Values and If/Else Statements
In Python, we use Booleans to tell the computer if something is true or not. Then, we use If/Else statements to tell the computer what to do next!

## ‚úÖ Boolean Values: True or False
A Boolean (bool) can only be one of two things: True or False.

Important: In Python, these must always start with a Capital Letter!

Example:

```python
is_it_sunny = True
is_it_raining = False

print(is_it_sunny)
print(type(is_it_raining)) # This will show <class 'bool'>
```

## ‚úÖ Comparison Operators: Asking Questions
We use special symbols to compare values. The result of these "questions" is always a Boolean (True or False).

### ‚öñÔ∏è Comparison Operators Table

| Symbol | Meaning | Example |
| :--- | :--- | :--- |
| `==` | Equal to | `5 == 5` is **True** |
| `!=` | Not equal to | `5 != 3` is **True** |
| `>` | Greater than | `10 > 5` is **True** |
| `<` | Less than | `2 < 1` is **False** |

## ‚úÖ The if Statement: Making Decisions
An if statement tells Python to run a block of code only if the condition is True.

Example:

```python
age = 10

if age >= 10:
    print("You are old enough to ride the roller coaster!")
```

### üí° Note: Don't forget the colon (:) at the end of the if line and the indent (space) before the print!

## ‚úÖ The else Statement: The "Otherwise" Plan
The else statement tells Python what to do if the if condition was False.

Example:

```python
is_hungry = False

if is_hungry == True:
    print("Time for a snack!")
else:
    print("Let's go play outside!")
```

## ‚ö†Ô∏è IMPORTANT: Indentation Matters!
In Python, the spaces at the start of the line (indentation) are how the computer knows which code belongs to the if or the else.

```python
if True:
    print("This is inside the if") # Indented!
print("This is outside the if")     # Not indented!
```
