### ‚úÖ Variable Property: They act as what they represent

This is the most important rule:

If a variable holds a number, it acts like a number.
If a variable holds a string, it acts like a string.

## üßÆ Variables with **Numbers**

Example:

```python
day = 1
week = day * 7
print(week)
```

üí° **Note:** Since day is a number, Python can do math with it!

## üßµ Variable with **Strings**

Remember: Variables act as what they represent.

So if a variable holds a string, it behaves like a string ‚úÖ

Example:

```python
name = "Chris"
print(name)
```

#### You can combine string variables using +

This is called concatenation (fancy word for **‚Äújoining‚Äù**).

```python
first = "Hello "
second = "World"
message = first + second
print(message)
```
