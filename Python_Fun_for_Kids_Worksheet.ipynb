## Python Building Blocks: Strings and Numbers


In Python, we mostly work with two basic types of information: **Strings** and **Numbers**.

### Strings
**Strings** are like words, sentences, or any text that Python needs to read. They *always* go inside quotation marks, either single (`'`) or double (`"`).

*   `"Hello"` is a string.
*   `'Python is dope!!!'` is a string.
*   `"12"` is also a string, even though it looks like a number, because it's inside quotes!

You can do cool things with strings:
*   **Add them together (concatenate):** `"Hello " + "World"` becomes `"Hello World"`
*   **Repeat them:** `"Repeat" * 3` becomes `"RepeatRepeatRepeat"`

### Numbers
**Numbers** are... well, numbers! They don't need quotation marks.

*   **Integers (int):** Whole numbers like `1`, `10`, `100`.
*   **Floats (float):** Numbers with a decimal point like `1.5`, `3.14`, `0.5`.

You can do math with numbers:
*   `5 + 2` (addition)
*   `10 - 3` (subtraction)
*   `4 * 2` (multiplication)
*   `9 / 3` (division)

**IMPORTANT:** You cannot mix strings and numbers directly with `+` or `*` unless you convert one of them. For example, `"My age is " + 10` will cause an error! But `"My age is " + "10"` works!


### Examples:

```
print("Hello World")
```

```
print('Python is fun')
```

```
print("Code" + "School")
```

```
print("Wow!" * 3)
```

```
print(5 + 7)
```

```
print(10 - 4)
```

```
print(2 * 6)
```

```
print(15 / 3)
```

```
# This will cause an error if uncommented: print("My age is " + 10)
```

## Remembering Things: Variables


Imagine you have a special box where you can store your favorite toys, and you put a label on the box, like "My Cars". In Python, we have something similar called a **variable**!

A **variable** is like a labeled container that holds information. You give it a name, and you put a value inside it using the `=` sign.

*   `age = 10`  (Here, `age` is the variable name, and `10` is the number it holds.)
*   `name = "Ally"` (Here, `name` is the variable, holding the string `"Ally"`.)

**Cool things about variables:**
*   They *remember* the value you give them.
*   You can use them in your code instead of typing the same value over and over.
*   They act like the thing they represent. If `age` is `10`, then `age + 5` is like `10 + 5`!

**Naming Rules for Variables:**
*   Can't start with a number (e.g., `1name` is bad).
*   Can't have spaces (e.g., `my age` is bad, use `my_age`).
*   Should be descriptive (e.g., `fruit_count` is better than `x`).


### Examples:

```
my_favorite_number = 7
print(my_favorite_number)
```

```
greeting = "Hello"
person = "World"
message = greeting + ", " + person + "!"
print(message)
```

```
x = 5
y = 3
sum_of_numbers = x + y
print(sum_of_numbers)
```

```
dog_age_factor = 7
my_age = 10
my_dog_age = my_age * dog_age_factor
print(f"My age in dog years is: {my_dog_age}")
```

## Getting Input: The input() Function


What if you want your Python program to ask a user for information, like their name or their favorite color? That's where the `input()` function comes in!

## Changing Types: Conversion Functions (int(), float(), str())


Remember how `input()` always gives you back a string, even if the user types a number? Sometimes we need to change the *type* of data so Python knows how to use it correctly. This is called **type conversion**.

We have three main magic functions for this:

*   `int()`: Changes something into a whole number (an **integer**).
    *   `int("5")` becomes the number `5`
    *   `int(3.14)` becomes the number `3` (it chops off the decimal part!)
*   `float()`: Changes something into a decimal number (a **float**).
    *   `float("2.5")` becomes the number `2.5`
    *   `float(7)` becomes the number `7.0`
*   `str()`: Changes something into a **string** (text).
    *   `str(123)` becomes the string `"123"`
    *   `str(4.5)` becomes the string `"4.5"`

**When do we use this?**
Often, when we get a number from `input()` and want to do math with it:
`age_str = input("How old are you? ")` # `age_str` is `"10"` (a string)
`age_num = int(age_str)` # `age_num` is `10` (a number!)
`print("Next year you will be " + str(age_num + 1))` # Convert back to string for printing with other strings!


### Examples:

```
# String to Number (Integer)
num_as_str = "10"
num_as_int = int(num_as_str)
print(f"'10' as integer: {num_as_int}")
print(f"Is it really a number? {num_as_int + 5}") # Try adding to it!

# String to Number (Float)
decimal_as_str = "3.14"
decimal_as_float = float(decimal_as_str)
print(f"'3.14' as float: {decimal_as_float}")
print(f"Is it really a decimal? {decimal_as_float * 2}")

# Number to String
my_lucky_number = 7
lucky_number_as_str = str(my_lucky_number)
print(f"7 as string: '{lucky_number_as_str}'")
print(f"Can we add it to text? " + lucky_number_as_str + " is my number.")
```
